```python
from sklearn.preprocessing import LabelEncoder
def encode_data(df):
    yes_no_map = {'no': 0, 'yes': 1}
    df['dependency'] = df['dependency'].replace(yes_no_map).astype(np.float32)
    df['edjefe'] = df['edjefe'].replace(yes_no_map).astype(np.float32)
    df['edjefa'] = df['edjefa'].replace(yes_no_map).astype(np.float32)
    df['idhogar'] = LabelEncoder().fit_transform(df['idhogar'])
```

```python
encode_data(train)
encode_data(test)
```

```python
def do_features(df):
    # the relation btween the two columns  can be getten by divition 
    feats_div = [('children_fraction', 'r4t1', 'r4t3'),
                  ### r4t1, persons younger than 12 years of age
                  ### r4t3,Total persons in the household
                 ('working_man_fraction', 'r4h2', 'r4t3'),
                 #### Males 12 years of age and older
                 #### r4t3,Total persons in the household
                 ('all_man_fraction', 'r4h3', 'r4t3'),
                 ### r4h3, Total males in the household
                 ### r4t3,Total persons in the household
                 ('human_density', 'tamviv', 'rooms'),
                 ## tamviv,  number of persons living in the household
                 ## rooms,  number of all rooms in the house
                 ('human_bed_density', 'tamviv', 'bedrooms'),
                  ## tamviv,  number of persons living in the household
                  ### bedrooms, number of bedrooms
                 ('rent_per_person', 'v2a1', 'r4t3'),
                 ### v2a1, Monthly rent payment
                 ### r4t3,Total persons in the household
                 ('rent_per_room', 'v2a1', 'rooms'),
                 ### v2a1, Monthly rent payment
                 ## rooms,  number of all rooms in the house
                 ('mobile_density', 'qmobilephone', 'r4t3'),
                 ### qmobilephone, # of mobile phones
                 ### r4t3, Total persons in the household
                 ('tablet_density', 'v18q1', 'r4t3'),
                 ###r4t3, Total persons in the household
                 ('mobile_adult_density', 'qmobilephone', 'r4t2'),
                 ('tablet_adult_density', 'v18q1', 'r4t2'),
                 ###r4t2, persons 12 years of age and older 
                ]
    
    
    # the relation btween the two columns  can be getten by subtraction 
    feats_sub = [('people_not_living', 'tamhog', 'tamviv'),
                 #### tamhog,  size of the household
                 #### tamviv, number of persons living in the household
                ]
    
    
    for f_new, f1, f2 in feats_div:
        df['fe_' + f_new] = (df[f1] / df[f2]).astype(np.float32)       
    for f_new, f1, f2 in feats_sub:
        df['fe_' + f_new] = (df[f1] - df[f2]).astype(np.float32)
        
    # Drop id's
    
    df.drop(['Id', 'idhogar'], axis=1, inplace=True)
    # Drop repeated columns
    df.drop(['hhsize', 'female', 'area2'], axis=1, inplace=True)
    
    
    return df
```

```python
train = do_features(train)
test = do_features(test)
```

```python
train = train.fillna(0)
test = test.fillna(0)
```

```python
X = train.drop(columns=["Id","Target"], axis=1, errors='ignore')
y = train["Target"].values
```

```python
def get_lgb_model():
    lgb_model = lgb.LGBMClassifier(boosting_type='dart', learning_rate=0.2,objective='multiclassova', class_weight='balanced', random_state=0)
    return lgb_model
```

```python
lgb_model = get_lgb_model()
lgb_model.fit(X, y)
target_lgb = lgb_model.predict(test)
```
